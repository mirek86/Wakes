```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.integrate import quad

sigma_z = 20e-6 # bunch lenght
x = np.linspace(-6*sigma_z, 6*sigma_z, 1000)

def gaussian(x, sigma):
    c = np.sqrt(2 * np.pi)
    return np.exp(-0.5 * (x / sigma)**2) / sigma / c

sigma1, sigma2, sigma3 = 1, 1.5,  2
g1, g2, g3 = gaussian(x, sigma_z), gaussian(x, sigma2), gaussian(x, sigma3)

plt.plot(x, g1)
#plt.plot(x, g2)
#plt.plot(x, g3)

plt.show()

quad(gaussian, -4*sigma_z, 4*sigma_z, args=(sigma_z/6))
```
